```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as srn
```

```python
from datetime import datetime
```

```python
data = pd. read_csv("USvideos.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data=data.drop_duplicates()
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
columns_to_remove = ['thumbnail_link','description']
data = data.drop(columns=columns_to_remove)
data.info()
```

```python
data["trending_data"]= data["trending_date"].apply(lambda x: datetime.strptime(x,'%y.%d.%m'))
data.head(3)
```

```python
data['publish_time']=pd.to_datetime(data['publish_time'])
data.head(2)
```

```python
data['publish_month']=data['publish_time'].dt.month
data['publish_day']=data['publish_time'].dt.day
data['publish_hour']=data['publish_time'].dt.hour
data.head(2)
```

```python
print(sorted(data["category_id"].unique()))
```

```python
data['category_name']= np.nan
```

```python
data.loc[(data["category_id"]== 22), "category_name"] = 'File and Animation'
data.head()
```

```python
data['year']= data['publish_time'].dt.year
yearly_counts= data.groupby('year')['video_id'].count()
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total publish count', title='Total publish video per year')
plt.show()
```

```python
yearly_views= data.groupby('year')['views'].sum()
yearly_views.plot(kind='bar', xlabel='Year', ylabel='Total views', title='Total views per year')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
category_views= data.groupby('category_id')['views'].sum().reset_index()
# it's different from that video bc i did not change the category name 
top_categories = category_views.sort_values(by='views', ascending=False).head(5)
plt.bar(top_categories['category_id'],top_categories['views'])
plt.title ( "Top 5 category", fontsize=15 )
plt.xlabel('category name', fontsize=12)
plt.ylabel('Top views' ,fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
srn.countplot(x='category_id', data=data,order=data['category_id'].value_counts().index)
plt.xticks(rotation=90)
plt.title('video count by category')
plt.show()
```

```python
video_per_hr= data['publish_hour'].value_counts().sort_index()
plt.figure(figsize=(12,6))
srn.barplot(x=video_per_hr.index,y=video_per_hr.values,palette='rocket')
plt.xticks(rotation=45)
plt.xlabel('hour of day', fontsize=12)
plt.ylabel('no of videos' ,fontsize=12)
plt.title('No of videos published per hour')
plt.show()
```

```python
data['publish_time']= pd.to_datetime(data['publish_time'])
data['publish_date']=data['publish_time'].dt.date
video_count_by_date=data.groupby('publish_date').size()
plt.figure(figsize=(12,6))
srn.lineplot(data=video_count_by_date)
plt.xticks(rotation=45)
plt.xlabel('publish date', fontsize=12)
plt.ylabel('no of videos' ,fontsize=12)
plt.title('No of videos published over time')
plt.show()
```

```python
srn.scatterplot(data=data, x='views',y='likes')
plt.xlabel('views', fontsize=12)
plt.ylabel('like' ,fontsize=12)
plt.title('views vs like')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplots_adjust(wspace=0.2,hspace=0.4,top=0.9)
plt.subplot(2,2,1)
g=srn.countplot(x='comments_disabled',data=data)
g.set_title("Coments Disabled")
plt.subplot(2,2,2)
g1=srn.countplot(x='ratings_disabled',data=data)
g1.set_title("Rating disabled")
plt.subplot(2,2,3)
g1=srn.countplot(x='video_error_or_removed',data=data)
g1.set_title("video error or disabled")
plt.show()
```
